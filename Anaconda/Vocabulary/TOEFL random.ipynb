```python
df1=pd.read_excel("TOEFL vocabulary.xlsx")
df1=df1.rename(columns={'Unnamed: 10':'长度'})
df1.head(15)
```

```python
import random
indices=[]
cnt=0
while cnt<100:
    idx=random.randrange(1,len(df1))
    if idx not in indices:
        indices.append(idx)
        cnt+=1
voc=df1.loc[indices,['词条','含义']]
pd.set_option('display.max_rows',100)
print('\n'.join(sorted(voc['词条'])))
```

```python
print(voc.sort_values('词条'))
```
