### pearson correlation coefficient

```python
df1 = pd.read_csv('Relevance.csv')
df1.head()
```

```python
plt.figure(figsize=(10,10))
train_corr=df1.corr()
sns.heatmap(train_corr,vmax=1,square=True,fmt=".3f",annot=True)
plt.savefig('C:/Users/dell/Desktop/Python Data Analysis/Relevance1.tif',
           dpi=600,
           bbox_inches='tight',
           facecolor='w',
           edgecolor='b')
plt.show()
```

```python
plt.figure(figsize=(10,10))
train_corr=df1.corr(method="spearman")
sns.heatmap(train_corr,vmax=1,square=True,fmt=".3f",annot=True)
plt.savefig('C:/Users/dell/Desktop/Python Data Analysis/Relevance2.tif',
           dpi=600,
           bbox_inches='tight',
           facecolor='w',
           edgecolor='b')
plt.show()
```

```python
plt.figure(figsize=(10,10))
train_corr=df1.corr()
sns.clustermap(train_corr,vmax=1,square=True,fmt=".3f",annot=True)
plt.savefig('C:/Users/dell/Desktop/Python Data Analysis/Relevance1.tif',
           dpi=600,
           bbox_inches='tight',
           facecolor='w',
           edgecolor='b')
plt.show()
```

```python
plt.figure(figsize=(24,8))
plt.subplot(131)
corr1=df1.corr(method="pearson")
sns.heatmap(corr1,vmax=1,square=True,fmt=".2f",annot=True)
plt.subplot(132)
corr2=df1.corr(method="spearman")
sns.heatmap(corr2,vmax=1,square=True,fmt=".2f",annot=True)
plt.subplot(133)
corr3=df1.corr(method="kendall")
sns.heatmap(corr3,vmax=1,square=True,fmt=".2f",annot=True)
plt.savefig('C:/Users/dell/Desktop/Python Data Analysis/Relevance3.tif',
           dpi=600,
           bbox_inches='tight',
           facecolor='w',
           edgecolor='b')
plt.show()
```

### max and min -imize

$x^\prime=\frac{x-min(x)}{max(x)-min(x)}$

```python
import sklearn.preprocessing as sk
df2=df1.sort_values(by='Total Score',ascending=False).drop('Marble',axis=1)
columns1=df2.columns
s1=sk.MinMaxScaler().fit_transform(df2) # No standardizing texts
s2=pd.DataFrame(s1,columns=columns1)
s2.head()
```

### Normal

$z=\frac{x-\mu}{\sigma}$

```python
s1=sk.Normalizer(norm='l2').fit_transform(df1.drop('Marble',axis=1)) # No standardizing texts
s2=pd.DataFrame(s1,columns=columns1)
s2.head()
```

```python
#multiple
def CustomClassifier(value):
    if value<10:
        return 'Low'
    elif 10<=value<40:
        return 'Medium'
    else:
        return 'High'
    
def CustomClassifier2(value):
    if value<20:
        return 'Low'
    elif 20<=value<80:
        return 'Medium'
    else:
        return 'High'
    
df2['Total Score_1']=df2['Total Score'].apply(CustomClassifier2)
df2['#1_1']=df2['#1'].apply(CustomClassifier)
df2['#2_1']=df2['#2'].apply(CustomClassifier)
df2.head(50)
```
